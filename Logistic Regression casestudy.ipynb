```python
def findcorr(data):
    col = data.columns
    for i in col:
        for j in col:
            cr = df[[i,j]].corr().iloc[1,0]
            #print('corr b/w {0},{1} is {2}'.format(i,j,cr))
            if (cr > 0.85 and cr !=1) or (cr < -0.5): # as we have alreay check no two var have 1 correlation
                print('corr b/w {0},{1} is {2}'.format(i,j,cr))
```

```python
def select_col(data):
    col = []
    count = 0
    for i in data:
        if (i[-1]==1):
            col.append(i[0])
            count +=1
        else:
            pass
    print(count)
    return col
```
